# Basic plots
Plots help us visualize the distribution of a data set

## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
tips = sb.load_dataset('tips')
tips.head(1)
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sb.pairplot(tips,hue='sex',palette='coolwarm')
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
# a =  A Series, the series's name attribute will be used to label the data axis.
# kde - the blue line of how the distribution should be
# bins - the number of the bins (טווחים) in the table, for exmple 5 will be 1-11, 11-21, 21-31, 31-41, 41-51
sb.distplot(a = tips['total_bill'],kde=False,bins=5)
```

## boxplot

```python
sb.boxplot(tips['total_bill'])
```

# Heatmap

## Basic Heatmap

```python
flights = sb.load_dataset("flights") # Load the example flights dataset
flights.head(1)
```

```python
pivot_flights = flights.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(10, 7)) # figsize=(Width, Height)

# annot - <optional> To put the numbers inside the table
# fmt - <optional> String formatting code to use when adding annotations
# linewidths - <optional> Width of the lines that will divide each cell
# ax - <optional> matplotlib Axes in which to draw the plot
sb.heatmap(pivot_flights, annot=True, fmt="d", linewidths=0.5, ax=ax)
```

## Correlation Heatmap
To find correlation between the columns in the table.
If the annot number is negative so the corr between the columns is negative - as long as A is getting bigger, B is getting smaller.

```python
tips = sb.load_dataset("tips") # Load the example tips dataset
tips.head(1)
```

```python
corr = pd.get_dummies(tips).corr() # Convert categorical variable into dummy/indicator variables "Male" to 1 "Female" to -1
plt.figure(figsize=(16,9)) # figsize=(Width, Height)

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sb.heatmap(corr,annot=True,linewidth = 0.5, cmap='coolwarm', mask=mask)
```

## Heatmap of pivot table

```python
heatmap1_data = pd.pivot_table(tips, values='size', 
                     index=['day', 'time'], 
                     columns=['sex'])
sb.heatmap(heatmap1_data,annot=True,linewidth = 0.5, cmap='coolwarm')
```
